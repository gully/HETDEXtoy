```python
from astroML.linear_model import NadarayaWatson
model = NadarayaWatson('gaussian', 10.0)
model.fit(wl[:, None], fl, fl*0.002)
fl_pred = model.predict(wl[:, None])

fig = plt.figure(figsize=(10, 6))

ax = fig.add_subplot(111)
ax.plot(wl, fl_pred, '-b', linewidth = 1, 
        label='Observed')
ax.set_xlim((350,550))
ax.set_xlabel('$\lambda$ (nm)')
ax.set_ylabel('$f$ (photons s$^{-1}$ m$^{-2}$ $\mu$m$^{-1}$ arcsec$^{-2}$)')
plt.show()
```
